## For the top 10 user rated books 

```python
user_rate = Ratings_users.merge(Book,on = 'ISBN') #Here we are merging the book database with pic with the user ratings dataset 
```

```python
num_rates = user_rate.groupby('Book-Title').count()['Book-Rating'].reset_index() #Here book rating stands for number of ratings
num_rates
```

```python
avg_rates = user_rate.groupby('Book-Title').mean()['Book-Rating'].reset_index() 
avg_rates.rename(columns = {'Book-Rating' : 'avg_rating'}, inplace=True) #Here we finding the average number of ratings 
avg_rates
```

```python
popul = avg_rates.merge(num_rates,on = 'Book-Title')
popul
```

```python
top_20 = popul[popul['Book-Rating']>=200].sort_values('avg_rating',ascending=False).head(20)
top_20
```

```python
pop = top_20.merge(Book ,on = 'Book-Title').drop_duplicates('Book-Title')
```

```python
pop = pop[['Book-Title','Book-Rating','Book-Author','Image-URL-M']]
pop
```
