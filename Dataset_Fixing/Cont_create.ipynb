```python
import numpy as np
import matplotlib.pyplot as mtp
import pandas as pd 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
books=pd.read_csv("new_books.csv")
books
```

```python
books= books[['genres','Book-Title','author','series','bookId','Book-Author','description','ISBN_y','Image-URL-M']]
```

```python
for i in range (0,27191):
    books['author'][i] = books['Book-Author'][i].replace(" ","")
```

```python
books
```

```python
for i in range (0,27191):
    books['author'][i] = books['author'][i].split()
```

```python
books
```

```python
for i in range(0,27191):
    a = books['genres'][i].split()
    books['genres'][i] = [b.strip(".,!;()[]'") for b in a]
```

```python
for i in range(0,27191):
    try:
        a = books['series'][i].split()
        books['series'][i] = [b.strip(".,!;()[]'") for b in a]
    except:
        continue
```

```python
books['tags'] = books['genres'] + books ['series']
```

```python
final_df = books[['Book-Title','Book-Author','description','ISBN_y','Image-URL-M','tags']]
```

```python
final_df.to_csv("final_ds.csv")
```
