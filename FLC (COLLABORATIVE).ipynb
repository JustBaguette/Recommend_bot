 ## popularity based recommendation


```python
ratings=rating.merge(book,on='ISBN')
```

```python
num_rating_df=ratings.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df=ratings.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg _ratings'},inplace=True)
avg_rating_df
```

```python
popular_df= num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df= popular_df[popular_df['num_ratings']>=250].sort_values('avg _ratings',ascending=False).head(50)
```
